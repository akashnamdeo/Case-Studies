# 3. Visual Analysis - Univariate & Bivariate

    For continuous variable(s): Distplot, countplot, histogram for univariate analysis 
    For categorical variable(s): Boxplot 
    For correlation: Heatmaps, Pairplots


```python
def getDistplot(df,colName):
    plt.figure(figsize=(20,10))
    sns.displot(x=colName,hue="Product",data=df,kde=True)
    plt.grid()
    plt.title("Distribution of count by {}".format(colName))
    plt.show()

def getcountplot(df,colName):
    plt.figure(figsize=(20,10))
    sns.countplot(x=df[colName],data=df,order=df[colName].value_counts().index,hue="Product")
    plt.xticks(rotation=90)
    plt.grid()
    plt.title("Count by {}".format(colName))
    plt.show()
    
def gethistplot(df,colName,bins="auto"):
    plt.figure(figsize=(20,10))
    sns.histplot(x=colName,data=df,bins=bins,hue="Product")
    plt.xticks(rotation=90)
    plt.grid()
    plt.show()
```

## Outlier Detection

The usual definition of outlier is data outside range Q1-1.5*IQR and Q3+1.5*IQR

```python
outlierCols=["Age","Education","Income","Miles"]
def plotBoxPlot(df,colName):
    sns.boxplot(x=df[colName])
    plt.show()

def OutlierDetection(df,colName):
    Q3=np.percentile(df[colName],75)
    IQR=np.percentile(df[colName],75)-np.percentile(df[colName],25)
    return df[df[colName]>Q3+1.5*IQR][colName].values

for i in outlierCols:
    plotBoxPlot(df,i)
    print("Outliers for {} Feature :".format(i),OutlierDetection(df,i))
```
